Matplotlib is a comprehensive library for creating static, animated, and interactive visualizations in Python. Matplotlib makes easy things easy and hard things possible.


*   Create publication quality plots.
*   Make interactive figures that can zoom, pan, update.
*   Customize visual style and layout.
*   Export to many file formats.
*   Embed in JupyterLab and Graphical User Interfaces.
*   Use a rich array of third-party packages built on Matplotlib.



```python
import matplotlib.pyplot as plt
```

```python
x = [1,2,3,4,5,6,7]
y = [50,51,52,48,47,49,46]
```

```python
# To simply plot this:
plt.plot(x,y)
```

```python
# You can google api's for plot and it has variety of functions
```

```python
# Color: sets the color of the line
plt.plot(x,y, color='pink')
```

```python
# linewidth: sets the width of the line
plt.plot(x,y,color='black', linewidth=8)
```

```python
# linestyle: sets like kinda the texture of the line
plt.plot(x,y,color='orange',linewidth=7,linestyle='dashdot')
```

```python
plt.plot(x,y,color='orange',linewidth=7,linestyle='dotted')
```

```python
# If you can see my chart doesn't have X and Y axis label, I can set that by:
# X = day
# Y = temperature
plt.xlabel('Day')
plt.ylabel('Temperature')

# Set the title of the chart
plt.title("Weather")

# Plotting the plot
plt.plot(x,y,color='orange',linewidth=2,linestyle='dotted')
```

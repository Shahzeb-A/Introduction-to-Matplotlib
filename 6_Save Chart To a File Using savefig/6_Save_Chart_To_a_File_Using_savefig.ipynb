```python
# In this tutorial we are going to learn how to save Matplotlib chart to a file
```

```python
import matplotlib.pyplot as plt
```

```python
value_expenses = [1400, 600, 300, 410, 250]
labels_expenses = ["Home Rent", 'Food', 'Phone/Internet bill','Car', 'Other utilities']

plt.axis('equal')
plt.pie(value_expenses,labels=labels_expenses, autopct='%.2f%%', radius = 1.4, shadow=True, explode=[0,0.1,0,0,0])
# For this you can you savefig function and in brackets supply the name of the file
plt.savefig("Chart2.png")
```

```python
value_expenses = [1400, 600, 300, 410, 250]
labels_expenses = ["Home Rent", 'Food', 'Phone/Internet bill','Car', 'Other utilities']

plt.axis('equal')
plt.pie(value_expenses,labels=labels_expenses, autopct='%.2f%%', radius = 1.4, shadow=True, explode=[0,0.1,0,0,0])
# If you use bbox inches with savefig, it's gonna fit the chart very well in the saved picture
plt.savefig("Chart3", bbox='tight')
```

```python
value_expenses = [1400, 600, 300, 410, 250]
labels_expenses = ["Home Rent", 'Food', 'Phone/Internet bill','Car', 'Other utilities']

plt.axis('equal')
plt.pie(value_expenses,labels=labels_expenses, autopct='%.2f%%', radius = 0.7, shadow=True, explode=[0,0.1,0,0,0])
# Sometimes you want to  have extra space in the picture for the chart like padding
# You can use another arguement called pad_inches, let's say I want 2 inch of padding
plt.savefig("Chart6", bbox='tight', pad_inches=2)
```

```python
value_expenses = [1400, 600, 300, 410, 250]
labels_expenses = ["Home Rent", 'Food', 'Phone/Internet bill','Car', 'Other utilities']

plt.axis('equal')
plt.pie(value_expenses,labels=labels_expenses, autopct='%.2f%%', radius = 0.7, shadow=True, explode=[0,0.1,0,0,0])

# You can also set the background to transparent or not
plt.savefig("Chart7.png", bbox='tight', pad_inches=2, transparent=True)
```

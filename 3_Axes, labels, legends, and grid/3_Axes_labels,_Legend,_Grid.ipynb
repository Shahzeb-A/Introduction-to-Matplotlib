```python
# In this tutorial we are going to look at Axes labels, Legends, and grid aspects of any 2D Plot
```

```python
import matplotlib.pyplot as plt
```

```python
# Let's use some static data

# Weekdays ( 1-7 )
days = [1,2,3,4,5,6,7]
# Maximum Temperature recorded in the weekdays
max_t = [50,51,52,48,47,49,46]
# Miniimum Temperature recorded in the weekdays
min_t = [43,42,40,44,33,35,37]
# Average Temperature recorded in the weekdays
avg_t = [45,48,48,46,40,42,41]
```

```python
# Plotting a chart for my maximum temperature
plt.plot(days,max_t)
```

```python
# In the same chart you can plot 3 different lines
plt.plot(days,max_t)
plt.plot(days,min_t)
plt.plot(days,avg_t)
```

```python
# Now when you look at the above chart, the things that is missing:
# 1. First of all I don't know what is x-axis and y-axis
# 2. I don't know what each of these lines represent
# 3. There is no title

# To set the x-axis and y-axis label, we can call the xlabel and ylabel method
plt.xlabel('Day')
plt.ylabel('Temperature')

plt.plot(days,max_t)
plt.plot(days,min_t)
plt.plot(days,avg_t)
```

```python
# You can also set the title using the title function
plt.title("Weather")

plt.xlabel('Day')
plt.ylabel('Temperature')

plt.plot(days,max_t)
plt.plot(days,min_t)
plt.plot(days,avg_t)
```

```python
# The chart is better than before but still it's not that good because I don't know what each of these lines represent so I need to have a legend here
# In order to have a legend, you need to first set a label on each othe plot function

plt.title("Weather")

plt.xlabel('Day')
plt.ylabel('Temperature')

plt.plot(days,max_t, label = "Maximum Temperature")
plt.plot(days,min_t, label = 'Minimumm Temperature')
plt.plot(days,avg_t, label = 'Average Temperature')

plt.legend()
```

```python
# Sometimes you want to change the location of the legend, you can do it using the loc property
plt.title("Weather")

plt.xlabel('Day')
plt.ylabel('Temperature')

plt.plot(days,max_t, label = "Maximum Temperature")
plt.plot(days,min_t, label = 'Minimumm Temperature')
plt.plot(days,avg_t, label = 'Average Temperature')

plt.legend(loc='upper right')
```

```python
plt.title("Weather")

plt.xlabel('Day')
plt.ylabel('Temperature')

plt.plot(days,max_t, label = "Maximum Temperature")
plt.plot(days,min_t, label = 'Minimumm Temperature')
plt.plot(days,avg_t, label = 'Average Temperature')

# You can also use the shadow property so it'll create a nice shadow behind the legend
plt.legend(loc='best', shadow=True)
```

```python
plt.title("Weather")

plt.xlabel('Day')
plt.ylabel('Temperature')

plt.plot(days,max_t, label = "Maximum Temperature")
plt.plot(days,min_t, label = 'Minimumm Temperature')
plt.plot(days,avg_t, label = 'Average Temperature')

# You can also control the font size of the legend using fontsize property
plt.legend(loc='upper right', shadow=True, fontsize=15)
```

```python
plt.title("Weather")

plt.xlabel('Day')
plt.ylabel('Temperature')

plt.plot(days,max_t, label = "Maximum Temperature")
plt.plot(days,min_t, label = 'Minimumm Temperature')
plt.plot(days,avg_t, label = 'Average Temperature')

plt.legend(loc='upper right', shadow=True)

# Another thing people like when doing 2D plotting is using grids, it can help visualize the data-points better
plt.grid()
# So whenever you're looking at any data-point you can easily locate them to x-axis using grid lines
```

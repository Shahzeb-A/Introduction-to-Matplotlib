```python
# We are going to be looking at format strings today which can be used to control visual aspects of your 2D plot such as color, marker, linestyle,etc
```

```python
import matplotlib.pyplot as plt
```

```python
x = [1,2,3,4,5,6,7]
y = [50,51,52,48,47,49,46]
```

```python
plt.plot(x,y)
```

```python
# You can control visual properties of this chart by specifying format strings as a third arguement
# Right now the line color is blue, let's say I want the marker to be specified g+
# Marker are the points on the graph
plt.plot(x,y,'g+')
# g means green
# + means marker
```

```python
# Check Matplotlib documentation for furthermore list format strings
```

```python
# If I want my linestyle to be dash dash, I can say --
# + are the points on the graph
# g is the color of the line
# -- is the linestyle
plt.plot(x,y,'g+--')
```

```python
# We can do many various styles
plt.plot(x,y,'g+-.')
```

```python
# Doesn't matter in which order you specify format string in, it'll work anyway
# So order doesn't matter
plt.plot(x,y,'+--b')
```

```python
# Instead of plus,  let's try diamond
plt.plot(x,y,'bD')
```

```python
# Now these might not be liked by some people, so you can also do it explicitly
# So we have acheived the same result as above but here we have mentioned it more clearly
plt.plot(x,y, color='blue', marker='D', linestyle='')
```

```python
# You can also specify any hexadecimal string instead of color
plt.plot(x,y, color='#008840', marker='D', linestyle='')
```

```python
# You can also control the marker size using the markersize arguement
plt.plot(x,y, color='#008840', marker='D', linestyle='', markersize=8)
```

```python
# You can also control the transparency of the line using alpha, ( the scale of alpha is between 0 and 1 )
plt.plot(x,y, color='#008840',alpha=0.3)
```
